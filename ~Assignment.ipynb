11. Ask user to enter the marks and define if the user got distinction, first class, second class, pass, Fails.

  If marks are more than 80 output must be “You got distinction”.

  For marks more than 60 output must be “You got First class”.

  Marks more than 40 will display “You got second class”.

  Marks more than or equal to 35 “PASS” Otherwise Fail.

```python
marks = input("Obtained marks : ")
if marks.isdigit():
  marks = int(marks)
  if 0 <= marks <= 100:
    if marks >= 80:
      print("You got Distinction")
    elif marks >= 60:
      print("You got First class")
    elif marks >= 40:
      print("You got Second class")
    elif marks >= 35:
      print("PASS")
    else:
      print("Fail")
  else:
    print("Enter valid marks between 0 and 100")
else:
  print("Enter valid marks")
```

14.	Digitalize the book allotment process for school. Charges are mentioned here in the given table:


```python
Books = {
    "Hindi": {'1-4':60,'5-8':100,'9-10':150},
    "Marathi": {'1-4':60,'5-8':100,'9-10':150},
    "English": {'1-4':80,'5-8':100,'9-10':150},
    "Science": {'1-4':90,'5-8':120,'9-10':200},
    "Maths": {'1-4':100,'5-8':140,'9-10':250}
}

Notebooks = {
    "square":{100:40,200:70},
    "4lines":{100:30,200:50},
    "2lines":{100:30,200:50},
    "singleline":{100:60,200:100},
    "a4":{100:100,200:180}
}

receipt = []
total_cost = 0
receipt_data_dict = {}

print("\n{:^150}".format("---Welcome to Book Store---"))

while True:

    print("\n{:^150}".format("What do you want to buy? (Book/Notebook): "))
    choice = input("\n{:^150}".format("->")).lower()

    # --------------for book-------------------
    if choice == "book":
        standard = input(("Enter class (1-4/5-8/9-10): "))
        subject = input("Enter subject \n(maths,marathi,hindi,science,english): ").capitalize()
        qty = int(input("Enter quantity: "))

        if subject in Books and standard in Books[subject]:
            price = Books[subject][standard]
            cost = price
            total_cost += price*qty
            receipt.append([subject + " Book", qty, cost, price*qty]) # Store subtotal for this item
        else:
            print("Invalid Book Details!")

    # ----------------for notebook----------------
    elif choice == "notebook":
        note_type = input("Enter notebook type \n(square,4lines,2lines,singleline,a4): ").lower()
        pages = int(input("Enter pages (100/200): "))
        qty = int(input("Enter quantity: "))

        if note_type in Notebooks and pages in Notebooks[note_type]:
            price = Notebooks[note_type][pages]
            cost = price
            total_cost += price*qty
            receipt.append([note_type + " Notebook", qty, cost, price*qty]) # Store subtotal for this item
        else:
            print("Invalid Notebook Details!")

    else:
        print("Invalid Choice!")

    # ----------------does user wants more entries----------------
    more = input("\nDo you want anything more? (Y/N): ").upper()
    if more == 'N':
        break
    if more == 'Y':
        continue
    else :
      print('invalid response')



for idx, item_details in enumerate(receipt):
    receipt_data_dict[idx + 1] = {
        'item': item_details[0],
        'quantity': item_details[1],
        'unit_price': item_details[2],
        'subtotal': item_details[3]
    }


# ----------------bill printing----------------

print('-'*60)
print("|{:^54}{:>6s}".format("WELCOME TO BOOK STORE","|"))
print("|{:^54}{:>6s}".format("RECEIPT","|"))
print("|",'-'*57,"|")
print("|{:<5s} {:28s} {:>9s} {:>10s}{:>5s}".format("Sr","Item","Qty","Price","|"))

for e,i in enumerate(receipt):
    print("|{:<5} {:28s} {:>9} {:>10.2f}{:>5s}".format(e+1,i[0],i[1],i[2],"|"))

print("|",'-'*57,"|")
print("|{:>30s} {:>14.2f}{:>15s}".format("Total Cost", total_cost, "|"))
print("|",'-'*57,"|")

print("\nThank You for Shopping!")
```

1.Take input from user, ask which books you want to buy or which notebooks he wants make dictionary of that order

```python
Books = {
    "Hindi": {'1-4':60,'5-8':100,'9-10':150},
    "Marathi": {'1-4':60,'5-8':100,'9-10':150},
    "English": {'1-4':80,'5-8':100,'9-10':150},
    "Science": {'1-4':90,'5-8':120,'9-10':200},
    "Maths": {'1-4':100,'5-8':140,'9-10':250}
}

Notebooks = {
    "square":{100:40,200:70},
    "4lines":{100:30,200:50},
    "2lines":{100:30,200:50},
    "single lines":{100:60,200:100},
    "a4":{100:100,200:180}
}

receipt = {}
total_cost = 0

print("\n===== Welcome to Book Store =====")

while True:

    choice = input("\nWhat do you want to buy? (Book/Notebook): ").lower()

    # ---------------- BOOK ----------------
    if choice == "book":
        standard = input("Enter class (1-4/5-8/9-10): ")
        subject = input("Enter subject: ").capitalize()
        qty = int(input("Enter quantity: "))

        if subject in Books and standard in Books[subject]:

            unit_price = Books[subject][standard]
            item_total = unit_price * qty
            item_key = subject + " Book"

            total_cost += item_total

            receipt[item_key] = {
                "standard": standard,
                "qty": qty,
                "unit_price": unit_price,
                "total": item_total
            }

        else:
            print("Invalid Book Details!")

    # ---------------- NOTEBOOK ----------------
    elif choice == "notebook":
        note_type = input("Enter notebook type: ").lower()
        pages = int(input("Enter pages (100/200): "))
        qty = int(input("Enter quantity: "))

        if note_type in Notebooks and pages in Notebooks[note_type]:

            unit_price = Notebooks[note_type][pages]
            item_total = unit_price * qty
            item_key = note_type.capitalize() + " Notebook"

            total_cost += item_total

            receipt[item_key] = {
                "pages": pages,
                "qty": qty,
                "unit_price": unit_price,
                "total": item_total
            }

        else:
            print("Invalid Notebook Details!")

    else:
        print("Invalid Choice!")

    more = input("\nDo you want anything more? (Y/N): ").upper()
    if more != 'Y':
        break


# ----------- SHOW STORED DICTIONARY -----------
print("\nStored Receipt Dictionary:\n")
print(receipt)


# ---------------- PRINT RECEIPT ----------------
print("-"*85)
print("|{:^83}|".format("WELCOME TO BOOK STORE"))
print("|{:^83}|".format("RECEIPT"))
print("-"*85)

print("|{:^4}|{:^25}|{:^12}|{:^8}|{:^15}|".format(
    "Sr","Item","Std/Page","Qty","Total"))
print("-"*85)

for index, (item, details) in enumerate(receipt.items(), start=1):

    if "standard" in details:
        extra = details["standard"]
    else:
        extra = str(details["pages"])

    print("|{:^4}|{:^25}|{:^12}|{:^8}|{:^15.2f}|".format(
        index,
        item,
        extra,
        details["qty"],
        details["total"]
    ))

print("-"*85)
print("|{:>65} {:>15.2f} |".format("Grand Total", total_cost))
print("-"*85)

print("\nThank You for Shopping! ")
```

37.	Create a function to accept marks from user utilize exception concept to validate proper marks.

```python
def check_palindrome(s):
    s = s.upper()
    reversed_string = ""

    for i in s:
        reversed_string = i + reversed_string

    if s == reversed_string:
        print("It is a Palindrome")
    else:
        print("It is Not a Palindrome")

check_palindrome("NITIN")
```

50.	Write a code to generate factorial of the number  For example: factorial of 5 = 5! = 5*4*3*2*1

```python
def factorial(n):
    fact = 1
    expression = ""

    for i in range(n, 0, -1):
        fact *= i
        expression += str(i)
        if i != 1:
            expression += " × "

    print(f"{n}! = {expression} = {fact}")

factorial(5)
```

52.	Write a program to check frequency of each element in the list.

```python
# Program to check frequency of each element in a list

def check_frequency(lst):
    freq = {}

    for item in lst:
        if item in freq:
            freq[item] += 1
        else:
            freq[item] = 1

    return freq


numbers = [1, 2, 2, 3, 4, 1, 2, 5, 3]

result = check_frequency(numbers)

for key, value in result.items():
    print(key, "appears", value, "times")
```
