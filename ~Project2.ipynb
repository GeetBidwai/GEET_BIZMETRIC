Ask user to opt for courses for master degree based on the following

L1 = [“HR”, “Finance”, “Marketing”, “DS”]

Based on above subject there are two different streams.

For example- HR is having HR core and HR analytics.

Marketing is having core and Marketing analytics.

Analytics is the optional subject and having added extra fees.

DS is not having analytics.

If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.

If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 .

Transportation charges 13000 per semester. Calculate the total annual cost based on selected service.

User will enter values as subject, analytics(Y/N), Hostel (Y/N), food(How many months?), Transportation(semester/annual)

```python
# by using if else:

L1=["HR","Finance","Marketing","DS"]
fees=200000
total=0

subject=input("Enter a subject:")
analytics=input("Is it analytics related(Yes/No) : ")
hostel=input("Are u living in hostel(yes/No): ")
food=int(input("Food: how many months? :"))
transport=input("Transportation(semester/annual) :")

if subject in L1:
    total=total+fees
    if analytics=="Yes":
      if subject != "DS":
        total=total+fees*0.10
      else:
        print("Analytics not available for DS.")

    if hostel=="Yes":
      total=total+200000

    total=total+2000*food

    if transport == "semester":
      total=total+13000
    elif transport == "annual":
      total=total+13000 * 2

    print("Total Annual Cost=",total)

else:
  print("Invalid Subject")
```

```python
# using dict

courses = {
    "HR": 200000,
    "Finance": 200000,
    "Marketing": 200000,
    "DS": 200000
}

transport_fees = {
    "semester": 13000,
    "annual": 26000
}

subject = input("Enter a subject: ")

if subject in courses:

    total = courses[subject]

    analytics = input("Analytics (Yes/No): ")
    if analytics == "Yes" and subject != "DS":
        total = total + courses[subject] * 0.10

    hostel = input("Hostel (Yes/No): ")
    if hostel == "Yes":
        total = total + 200000

    food = int(input("Food months: "))
    total = total + 2000 * food

    transport = input("Transport (semester/annual): ")
    if transport in transport_fees:
        total = total + transport_fees[transport]

    print("Total Annual Cost =", total)

else:
    print("Invalid Subject")
```

```python
# using try and except

L1 = ["HR", "Finance", "Marketing", "DS"]
fees = 200000

while True:
    subject = input("Enter a subject: ")
    if subject in L1:
        break
    else:
        print("Invalid Subject. Try again.")

analytics = input("Analytics (Yes/No): ")

hostel = input("Hostel (Yes/No): ")

while True:
    try:
        food = int(input("Food months: "))
        if food < 0:
            print("Food months cannot be negative.")
        else:
            break
    except ValueError:
        print("Please enter a valid number.")

while True:
    transport = input("Transport (semester/annual): ")
    if transport in ["semester", "annual"]:
        break
    else:
        print("Invalid transport option. Try again.")

total = fees

if analytics == "Yes" and subject != "DS":
    total += fees * 0.10

if hostel == "Yes":
    total += 200000

total += 2000 * food

if transport == "semester":
    total += 13000
else:
    total += 26000

print("Total Annual Cost =", total)
```

```python
# using OOPs

class Course:

    Fees = 200000

    def __init__(self, subject, analytics, hostel, food, transport):
        self.subject = subject
        self.analytics = analytics
        self.hostel = hostel
        self.food = food
        self.transport = transport

    def calculate(self):

        L1 = ["HR", "Finance", "Marketing", "DS"]

        if self.subject not in L1:
            return "Invalid Subject"

        total = self.Fees

        if self.analytics == "Yes" and self.subject != "DS":
            total += self.Fees * 0.10

        if self.hostel == "Yes":
            total += 200000

        total += 2000 * self.food

        if self.transport == "semester":
            total += 13000
        elif self.transport == "annual":
            total += 26000

        return total


subject = input("Enter subject: ")
analytics = input("Analytics (Yes/No): ")
hostel = input("Hostel (Yes/No): ")
food = int(input("Food months: "))
transport = input("Transport (semester/annual): ")

student = Course(subject, analytics, hostel, food, transport)
print("Total Annual Cost =", student.calculate())
```
