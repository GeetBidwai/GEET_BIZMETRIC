12. Use lambda to compute factorial using reduce() (yeah, that one-liner madness).

```python
from functools import reduce

def factorial(n):
  if n == 0:
    return 1
  else:
    return reduce(lambda x, y: x * y, range(1, n + 1))

print(f"Factorial of 5: {factorial(5)}")
print(f"Factorial of 0: {factorial(0)}")
print(f"Factorial of 1: {factorial(1)}")
print(f"Factorial of 7: {factorial(7)}")
```

19. Use lambda to calculate the area of a circle (πr²).

```python
import math

calculate_circle_area = lambda r: math.pi * (r**2)

radius1 = 5
area1 = calculate_circle_area(radius1)
print(f"The area of a circle with radius {radius1} is: {area1:.2f}")

radius2 = 10
area2 = calculate_circle_area(radius2)
print(f"The area of a circle with radius {radius2} is: {area2:.2f}")
```

20. Write a lambda to remove duplicates from a list using filter + set.

```python
original_list = [1, 2, 2, 3, 4, 4, 5, 1, 6, 7, 7]

seen = set()


remove_duplicates_lambda = lambda lst: list(filter(lambda x: x not in seen and not seen.add(x), lst))

unique_list = remove_duplicates_lambda(original_list)
print(unique_list)
```

21. Use lambda with reduce() to find the product of all numbers in a list.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product_of_numbers = reduce(lambda x, y: x * y, numbers)
print(f"The product of numbers {numbers} is: {product_of_numbers}")

numbers_two = [10, 2, 3]
product_of_numbers_two = reduce(lambda x, y: x * y, numbers_two)
print(f"The product of numbers {numbers_two} is: {product_of_numbers_two}")
```

27. Write a lambda to check if two strings are anagrams.

```python
anagrams = lambda s1, s2: __builtins__.sorted(s1.lower().replace(" ", "")) == __builtins__.sorted(s2.lower().replace(" ", ""))


print(f"'listen' and 'silent' are anagrams: {anagrams('listen', 'silent')}")
print(f"'hello' and 'world' are anagrams: {anagrams('hello', 'world')}")
print(f"'Debit card' and 'Bad credit' are anagrams: {anagrams('Debit card', 'Bad credit')}")
print(f"'Astronomer' and 'Moon starer' are anagrams: {anagrams('Astronomer', 'Moon starer')}")
```
