```python
print("\n Welcome to SWAAD Restaurant ")
print("--------------------------------------")
print("Here’s our delicious menu:\n")

# ---------- Menu ----------
cursor.execute("SELECT ID, Items, Price FROM Resturant_Menu")
menu_items = cursor.fetchall()

for item in menu_items:
    print(f"{item[0]}. {item[1]}  -  Rs.{item[2]}")

print("--------------------------------------")

grand_total = 0
ordered_items = []


while True:
    try:
        print("\n Place Your Order")
        item_id = int(input("Enter Item ID: "))
        quantity = int(input("Enter Quantity: "))

        if quantity <= 0:
            print("Quantity must be greater than 0.")
            continue

    except ValueError:
        print("Please enter valid numeric values.")
        continue

    # Fetch item details
    cursor.execute(
        "SELECT Items, Price FROM Resturant_Menu WHERE ID = ?",
        (item_id,)
    )
    item = cursor.fetchone()

    if not item:
        print("Invalid Item ID. Please try again.")
        continue

    item_name, price = item
    total_price = price * quantity

    # Save order into database
    cursor.execute(
        "INSERT INTO Orders (item_id, quantity, price, total) VALUES (?, ?, ?, ?)",
        (item_id, quantity, price, total_price)
    )
    conn.commit()

    # Store locally for bill
    ordered_items.append((item_name, quantity, price, total_price))
    grand_total += total_price

    print(f" {item_name} added successfully!")

    more = input("\nWould you like to order anything else? (yes/no): ").lower()
    if more != "yes":
        break


# ---------- Bill Printing ----------
print("\n\n --------------- FINAL BILL ---------------")

for item in ordered_items:
    print(f"{item[0]}  |  Qty: {item[1]}  |  Rs.{item[2]}  |  Total: Rs.{item[3]}")

print("---------------------------------------------")
print(f" Grand Total: Rs.{grand_total}")
print("---------------------------------------------")
print(" Thank you for dining with SWAAD!")

conn.close()
```

## Book Store Bill Printing Assignmnet

```python
import pyodbc

# ----------- Database Connection -----------
conn = pyodbc.connect(
    "Driver={SQL Server};"
    "Server=GEET\SQLEXPRESS;"
    "Database=Book_Printing;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

print("\nWelcome to School Book Store")
print("----------------------------------------")

grand_total = 0
cart = []

while True:

    print("\n What would you like to buy?")
    print("Choose 1 for Books")
    print("Choose 2 for Notebooks")

    choice = input("Enter your choice (1 or 2): ")

    # --------------- For Book -------------------
    if choice == "1":

        standard = input("\nEnter Standard (1st-4th / 5th-8th / 9th-10th): ")

        cursor.execute(
            "SELECT book_id, subject, price FROM Books WHERE standard=?",
            (standard,)
        )
        books = cursor.fetchall()

        if not books:
            print(" No books available for this standard.")
            continue

        print("\n Available Books:")
        for b in books:
            print(f"ID: {b.book_id} | {b.subject} | Price: Rs.{b.price}")

        try:
            book_id = int(input("\nEnter Book ID: "))
            quantity = int(input("Enter Quantity: "))
        except ValueError:
            print(" Please enter valid numbers.")
            continue

        cursor.execute(
            "SELECT subject, price FROM Books WHERE book_id=?",
            (book_id,)
        )
        book = cursor.fetchone()

        if book:
            name, price = book
            total = price * quantity

            cursor.execute(
                "INSERT INTO Orders (item_name, quantity, price, total) VALUES (?, ?, ?, ?)",
                (name, quantity, price, total)
            )
            conn.commit()

            cart.append((name, quantity, price, total))
            grand_total += total

            print(f" {name} added to cart successfully!")
        else:
            print(" Invalid Book ID.")


    # ---------------------------- For Notebook ------------------------------
    elif choice == "2":

        cursor.execute(
            "SELECT notebook_id, type, pages, price FROM Notebooks"
        )
        notebooks = cursor.fetchall()

        print("\n Available Notebooks:")
        for n in notebooks:
            print(f"ID: {n.notebook_id} | {n.type} - {n.pages} pages | Price: Rs.{n.price}")

        try:
            nb_id = int(input("\nEnter Notebook ID: "))
            quantity = int(input("Enter Quantity: "))
        except ValueError:
            print(" Please enter valid numbers.")
            continue

        cursor.execute(
            "SELECT type, pages, price FROM Notebooks WHERE notebook_id=?",
            (nb_id,)
        )
        nb = cursor.fetchone()

        if nb:
            type_name, pages, price = nb
            name = f"{type_name} {pages} pages"
            total = price * quantity

            cursor.execute(
                "INSERT INTO Orders (item_name, quantity, price, total) VALUES (?, ?, ?, ?)",
                (name, quantity, price, total)
            )
            conn.commit()

            cart.append((name, quantity, price, total))
            grand_total += total

            print(f" {name} added to cart successfully!")
        else:
            print(" Invalid Notebook ID.")

    else:
        print(" Please choose 1 or 2 only.")

    more = input("\nWould you like to add more items? (yes/no): ").lower()
    if more != "yes":
        break


# ----------- Final Bill -----------
print("\n---------------------------------------")
print("                FINAL BILL            ")
print("---------------------------------------")
print("Item\t\tQty\tPrice\tTotal")
print("----------------------------------------")

for name, qty, price, total in cart:
    print(f"{name}\t{qty}\tRs.{price}\tRs.{total}")

print("----------------------------------------")
print(f"Grand Total:\t\t\tRs.{grand_total}")
print("---------------------------------------")
print(" Thank you for shopping with us!")
print("---------------------------------------")

conn.close()
```
